```python
import zipfile
import os

zip_files = [
    '/lakehouse/default/Files/medicare/Medicare_Part_D_Prescribers_by_Provider_and_Drug_2013.zip',
    '/lakehouse/default/Files/medicare/Medicare_Part_D_Prescribers_by_Provider_and_Drug_2014.zip'
]

raw_path = '/lakehouse/default/Files/medicare/raw'
os.makedirs(raw_path, exist_ok=True)

for zip_path in zip_files:
    if os.path.exists(zip_path):
        print(f'Extracting: {zip_path}')
        with zipfile.ZipFile(zip_path, 'r') as z:
            z.extractall(raw_path)
        print(f'Done: {zip_path}')
    else:
        print(f'NOT FOUND: {zip_path}')

print('Completed. Contents of raw:')
for f in os.listdir(raw_path):
    print(f'  {f}')

# CELL_END ********************
```
